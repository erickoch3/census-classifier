```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load the dataset
file_path = '../../data/census.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
df.head()
```

```python
# Display basic information about the dataset
df.info()

# Display summary statistics
df.describe(include='all')
```

```python
# Check for missing values
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Check for duplicate rows
duplicate_rows = df[df.duplicated()]
print(f'Number of duplicate rows: {duplicate_rows.shape[0]}')
```

```python
import textwrap

# Display data types and unique values for categorical features
for column in df.select_dtypes(include=['object']).columns:
    unique_values = df[column].unique()
    unique_values_str = ', '.join([str(category) for category in unique_values])
    wrapped_values = textwrap.fill(unique_values_str, width=80, subsequent_indent=' ' * 8)
    print(f'{column}: {df[column].nunique()} unique values')
    print(f'\t{wrapped_values}')
```
